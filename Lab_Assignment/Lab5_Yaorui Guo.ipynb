# 1. Run a multiple linear probability model (have at least 2 Xs in the model). Tell me how you think your independent variables will affect your dependent variable. Interpret your results. Were your expectations correct? Why or why not?

## The question is “How often do people find their work stressful".  The answer range is from 1 to 5, which represents always to never.

## The dependent variable is 'stress', and the independent variables I have chosen are five possible influencing factors: "xhaustn","age", "babies", "prestg80","income". 

## I speculate that stress at work rises with the number of children and prestige, and decreases with age and income. Stress at work will be higher when work makes people more exhausted.

```python
sub = g.dropna(subset = ["stress", "xhaustn","age", "babies", "prestg80","income","wrkstat"]) 
sub
```

```python
sub = g.dropna(subset = ["jobsat", "hrs1", "age", "finrela", "safehlth", "proudemp"]) 
sub
```

## I am going to recode to look variable 'stress' at if someone says "Always" and "Often" vs. everything else.

```python
conditions = [
    (sub['stress'] <= 2),
    (sub['stress'] > 2)]
choices = [1, 0]
sub['str'] = np.select(conditions, choices, default=np.nan)
```

```python
sub.str.value_counts()
```

## I reverse code these independent variable 'xhaustn' to make the interpretion easily.

```python
## reverse code variable 'xhaustn' - higher variable means feel more exhaused from work ##
sub['rxhaustn'] = 6 - sub.xhaustn
```

# I will subset LPM on only full-time working people

```python
lm1 = smf.ols(formula = 'str ~ rxhaustn + age + babies + prestg80 + income', 
              subset=(sub['wrkstat']==1), 
              data = sub).fit()
print (lm1.summary())
```

## This LPM result shows that the variables 'rxhaustn'  and 'prestg80' are statistically significant for the dependent variable 'stress', while the other three variables all have p-values much greater than 0.05, so they are not statistically significant to stress.

## For every category increase in feeling exhausted at work, a full-time employee is 0.206  points more likely to gain stress from work on average, net of prestg80 and other variables.. 

## Having a higher prestige, makes full-time employee 0.46 percentage points more likely to gain stress from work on average, controlling for level of feeling exhausted at work and other variables.

## This result is roughly in line with my expectations. People are more likely to be more stressed at work when they feel more exhausted at work, or are in positions of high prestige.

# 2. Run a multiple (binary) logistic model. (It can be the same as the above LPM or a new model.) If it is a new model, tell me how you think your independent variables will affect your dependent variable. Interpret your results in the logit scale. Were your expectations correct? Why or why not?

```python
logit1 = sm.formula.logit(formula = 'str ~ rxhaustn + age + babies + prestg80 + income', 
                          subset=(sub['wrkstat']==1), 
                          data = sub).fit()
print (logit1.summary()) 
```

## From logit model result, for every category increase in feeling exhausted at work, a full time employee increases his/her logit by 0.989 of having strees from work on average, net of prestg80 and other variables.

## Having higher prestige increases a full time employee’s log-odds on having strees from work by 0.022 on average, controlling for level of feeling exhausted at work and other variables.

# 4. Extra Credit: Get predicted probabilities from your logit model in Question 2 for some constellations of X values and interpret the results.

```python
def logit2prob (logit):
    odds = np.exp(logit)
    prob = odds / (1 + odds) 
    return(prob);


intercept = logit1.params.Intercept
b_rxhaustn = logit1.params.rxhaustn
b_age = logit1.params.age
b_babies = logit1.params.babies
b_prestg80 = logit1.params.prestg80
b_income = logit1.params.income
```

## I want to know if I will feel stressful about my job after graduation. I am going to find out the predicted probability of being stressed for a person who feel moderately exhausted from work, is 24 years old, has a relatively high prestige job of 50, no baby, and can earn the average income.

```python
logits_str = intercept + (3 * b_rxhaustn) + (24 * b_age) + (50 * b_prestg80) + (0 * b_babies) \
             + (sub.income.mean() * b_income)
logit2prob(logits_str)
```
